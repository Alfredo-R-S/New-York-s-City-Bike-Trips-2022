# Kepler.gl Map Instance

```python
# Average lat/lng for each station name
start_coords = df.groupby('start_station_name')[['start_lat', 'start_lng']].mean().reset_index()
end_coords = df.groupby('end_station_name')[['end_lat', 'end_lng']].mean().reset_index()
```

```python
# Merge back to original dataframe
df = df.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'])
df = df.merge(start_coords, on='start_station_name', how='left')
df = df.merge(end_coords, on='end_station_name', how='left')
```

```python
# Group by station names and average coords, then sum values
agg_df = df.groupby([
    'start_station_name', 'end_station_name',
    'start_lat', 'start_lng', 'end_lat', 'end_lng'
]).agg({'value': 'sum'}).reset_index()
```

```python
agg_df
```

```python
# Save file
agg_df.to_csv("cleaned_bike_routes.csv", index=False)
```

```python
# Create KeplerGl instance

m = KeplerGl(height = 700, data={"data_1": agg_df})
m
```
